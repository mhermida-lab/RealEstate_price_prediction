# Stage 3: Model Selection

In the last stage, I've added new colums with get_dummies to transform categorical features into vectors manageable by the models.

The objective for my model is to accuratelly predict appartments prices with a set of features defined in the second stage.

The metric to measure model performance will be mean squared error which is represented by the formula:

$$RMSE = \sqrt{\frac{\sum_{t=1}^n (\hat y_t - y_t)^2}{n}}$$

This metric shows the error, wich is the diference between the predicted price and the actual price of the apparments.

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error

pd.set_option('display.float_format', lambda x: '%.3f' % x)
```

```python
df = pd.read_csv('./final_dataset.csv')
print("This dataset has {} observations".format(df.shape[0]))
```

To get started I'm going to asign the target (ars_price) to a serie y and the rest of the features to the Dataframe X. Then I'm going to split the  `dataset` into a training dataset (80%) and a test dataset (20%) to train the model with the first one and then measure the accuracy of predictions with the second one.



```python
X = df.drop(['ars_price'], axis=1)
y = df['ars_price']
```

```python
a, b = X.iloc[:,0:5].to_numpy(), X.iloc[:,5:].to_numpy()
scaler = StandardScaler()
scaled_a = scaler.fit_transform(a)
scaled_X = np.concatenate((scaled_a, b), axis=1)
pd.DataFrame(scaled_X)
```

```python
minmax = MinMaxScaler()
minmax_a = minmax.fit_transform(a)
minmax_X = np.concatenate((minmax_a, b), axis=1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.10)
scaled_X_train, scaled_X_test, y_train, y_test = train_test_split(scaled_X, y, test_size = 0.10)
minmax_X_train, minmax_X_test, y_train, y_test = train_test_split(minmax_X, y, test_size = 0.10)
```

```python
pd.DataFrame(minmax_X_train)
```

## Random Forest




I'm going to fit the model with the training set and then predict with the test set to measure how close each prediction is to the real target value

```python
from sklearn.ensemble import RandomForestRegressor
rf_reg = RandomForestRegressor(n_jobs=-1)
rf_reg.fit(minmax_X_train, y_train)
```

```python
rf_pred = rf_reg.predict(minmax_X_test)
mse = mean_squared_error(y_test, rf_pred)
print("RMSE: ", np.sqrt(mse))
```

## Lasso Regression

```python
from sklearn.linear_model import Lasso
```

```python
l_reg = Lasso(normalize=True)
l_reg.fit(scaled_X_train, y_train)
```

```python
lreg_pred = l_reg.predict(scaled_X_test)
mse = mean_squared_error(y_test, lreg_pred)
print("RMSE: ", np.sqrt(mse))
```

## Feed Forward Neural Network

```python
from keras.layers import Dense, Activation
from keras.models import Sequential
from keras.optimizers import Adam
import keras
```

```python
model = Sequential()
model.add(Dense(61, input_dim=(62), activation="relu"))
model.add(Dense(30, activation="relu"))
model.add(Dense(30, activation="relu"))
model.add(Dense(10, activation="relu"))
model.add(Dense(1, activation="linear"))
```

```python
opt = Adam(learning_rate=0.01)
model.compile(optimizer=opt,
              loss='mae',
             metrics=['accuracy']
             )
```

```python
es = keras.callbacks.EarlyStopping(monitor='val_loss',
                              min_delta=0,
                              patience=10,
                              verbose=1, mode='auto')
reduce_lr = keras.callbacks.ReduceLROnPlateau(monitor='val_loss', factor=0.2,
                              patience=5, min_lr=0.001)
```

```python

model.fit(minmax_X_train, y_train, validation_data=(minmax_X_test, y_test), epochs=5000, batch_size=30000, callbacks=[es, reduce_lr])
```

```python
nn_pred = model.predict(minmax_X_test)
mse = mean_squared_error(y_test, nn_pred)
print("RMSE: ", np.sqrt(mse))
```

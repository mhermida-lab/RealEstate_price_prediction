# Stage 2: Feature Engineering

Since in the first stage I've already done some feature selection and feature engineering, now I'm just going to prepare the data to be able to be used by the ML algorithms. The goal is to transform any categorical data into a more suitable form. In ordert o accomplish that task I'll use OneHotEncoding transforming categories into vectors with binary data. 

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('./cleaned_dataset.csv')
```

#### I'm going to drop lat and long features as those are highly correlated to neighborhoods. I'll keep just neighborhood. Same thing with ars_m2 and ars_price. No need to keep them both.


```python
df.drop(['lat','lon'], axis=1, inplace=True)
```

```python
df.head()
```

#### Then I'll transform the categorical feature into a vector.

```python
df = pd.get_dummies(df, columns=['neighborhood'])
```

```python
df.to_csv('./final_dataset.csv', header=True, index=False)
```
